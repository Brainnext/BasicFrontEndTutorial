# Pandas for indexing and accessing data

```python
import matplotlib.pyplot as plt
import pandas as pd

all = pd.read_csv ('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
new = pd.read_csv ('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
old = new.head(20)
new = old.loc[(old.logS == -2.180) & (old.MolWt >= 133.405) | (old.MolLogP >= 2.61186)]

print (new)
plt.plot(new)
plt.show()
```

```python
a = old.loc[old.logS.isin([-2.180, -3.20, -3.59])]

print (a)
plt.plot(a)
plt.show()
```

```python
old.loc[old.logS.notnull()]
```
